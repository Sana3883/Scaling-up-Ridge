```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_execution_times(num_rows):
    
    mkl_df_Parcel = pd.read_csv('mkl_execution_times_Parcel.csv', header=None, nrows=num_rows)
    openblas_df_Parcel = pd.read_csv('openblas_execution_times_Parcel.csv', header=None, nrows=num_rows)

    # Extract MKL and OpenBLAS execution time 
    MKL_Parcel = mkl_df_Parcel.values.tolist()[0]
    OPENBLAS_Parcel = openblas_df_Parcel.values.tolist()[0]

   
    df_Parcel = pd.DataFrame({'MKL': MKL_Parcel, 'OPENBLAS': OPENBLAS_Parcel})

 
    mkl_df_roi = pd.read_csv('mkl_execution_times_ROI.csv', header=None, nrows=num_rows)
    openblas_df_roi = pd.read_csv('openblas_execution_times_ROI.csv', header=None, nrows=num_rows)

    # Extract MKL and OpenBLAS data from ROI files
    MKL_roi = mkl_df_roi.values.tolist()[0]
    OPENBLAS_roi = openblas_df_roi.values.tolist()[0]


    df_roi = pd.DataFrame({'MKL': MKL_roi, 'OPENBLAS': OPENBLAS_roi})

    
    fig, axs = plt.subplots(1, 2, figsize=(16, 6))

    # Plot Parcel data
    error_Parcel = [35] * len(df_Parcel)
    df_Parcel.plot.bar(ax=axs[0], fontsize=16, yerr=error_Parcel, rot=0, width=0.9)
    axs[0].set_title('Parcel-wise', fontsize=18)
    axs[0].set_xlabel('Number of threads', fontsize=16)
    axs[0].set_ylabel('Training time (Seconds)', fontsize=16)
    axs[0].legend(fontsize=14)

    # Plot ROI data
    error_roi = [35] * len(df_roi)
    df_roi.plot.bar(ax=axs[1], fontsize=16, yerr=error_roi, rot=0, width=0.9)
    axs[1].set_title('ROI-wise', fontsize=18)
    axs[1].set_xlabel('Number of threads', fontsize=16)
    axs[1].set_ylabel('Training time (Seconds)', fontsize=16)
    axs[1].legend(fontsize=14)

    plt.tight_layout()
    plt.show()

```

```python
#Subject6
plot_execution_times(1)
```
