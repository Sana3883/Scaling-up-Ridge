```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

mkl_execution_times = np.genfromtxt('mkl_execution_times_Parcel.csv', delimiter=',').tolist()
openblas_execution_times = np.genfromtxt('openblas_execution_times_Parcel.csv', delimiter=',').tolist()

num_processors = [1, 2, 4, 8, 16, 32]

# Calculate speed-up for each subject and BLAS library
mkl_speedup = [[mkl_execution_times[i][0] / time for time in mkl_execution_times[i]] for i in range(len(num_processors))]
openblas_speedup = [[openblas_execution_times[i][0] / time for time in openblas_execution_times[i]] for i in range(len(num_processors))]


mkl_color = list(mcolors.TABLEAU_COLORS.values())[0]
openblas_color = list(mcolors.TABLEAU_COLORS.values())[1]

# Plotting
plt.figure(figsize=(12, 8))

for i in range(len(num_processors)):
    
    plt.plot(num_processors, mkl_speedup[i], label='MKL' if i == 0 else None, marker='o', color=mkl_color)
    
    plt.plot(num_processors, openblas_speedup[i], label='OpenBLAS' if i == 0 else None, marker='o', color=openblas_color)
    
    plt.plot(num_processors, mkl_speedup[i], marker='o', color=mkl_color) if i != 0 else None
    
    plt.plot(num_processors, openblas_speedup[i], marker='o', color=openblas_color) if i != 0 else None

plt.xlabel('Number of Threads', fontsize=20)
plt.ylabel('Speed-Up', fontsize=20)
plt.legend(fontsize=12)
plt.grid(True)

# Increase the font size of numbers on the axes
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.show()
```
