```python
import pandas as pd
import matplotlib.pyplot as plt

def plot(sub):
   
    df = pd.read_csv('execution_time_MOR.csv')

    # Filter data for the specified sub
    sub_data = df[df['Sub'] == sub]

    # Extract threads data
    threads2 = [int(x) for x in sub_data['Threads 2'].iloc[0].split()]
    threads4 = [int(x) for x in sub_data['Threads 4'].iloc[0].split()]
    threads8 = [int(x) for x in sub_data['Threads 8'].iloc[0].split()]
    threads16 = [int(x) for x in sub_data['Threads 16'].iloc[0].split()]
    threads32 = [int(x) for x in sub_data['Threads 32'].iloc[0].split()]

    langs = ['2', '4', '6', '8']

    df_plot = pd.DataFrame({
        '#Dask_MKL threads=2': threads2,
        '#Dask_MKL threads=4': threads4,
        '#Dask_MKL threads=8': threads8,
        '#Dask_MKL threads=16': threads16,
        '#Dask_MKL threads=32': threads32
    }, index=langs)

    error2 = [50, 50, 50, 50]

    ax = df_plot.plot.bar(figsize=(22, 12), fontsize=40, yerr=error2, rot=0, width=0.9)
    
    plt.xlabel('Number of Dask Compute Nodes', fontsize=45)
    plt.ylabel('Training time (seconds)', fontsize=45)
    
    plt.legend(fontsize=30)
    
    plt.show()

```

```python
plot('Sub1')
plt.show()
```
